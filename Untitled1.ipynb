```python
import gradio as gr
import pandas as pd

def predict_loan(loan_amnt, int_rate, annual_inc, total_dti):
    # Create a DataFrame with all columns present in X_train, initialized to 0
    input_data = pd.DataFrame(0, index=[0], columns=X_train.columns)

    # Fill in the values from Gradio inputs
    input_data['loan_amnt'] = loan_amnt
    input_data['int_rate'] = int_rate
    input_data['annual_inc'] = annual_inc
    input_data['total_dti'] = total_dti

    prediction = model.predict(input_data)[0]

    if prediction == 1:
        return "Loan Likely to be Fully Paid ✅"
    else:
        return "High Risk of Default ❌"

interface = gr.Interface(
    fn=predict_loan,
    inputs=[
        gr.Number(label="Loan Amount"),
        gr.Number(label="Interest Rate"),
        gr.Number(label="Annual Income"),
        gr.Number(label="Debt to Income Ratio")
    ],
    outputs="text",
    title="Fintech Loan Risk Prediction System"
)

interface.launch()
```
